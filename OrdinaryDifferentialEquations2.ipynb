```python
from CommonFunctions import *
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
```

# $xy' - y = x^2 sin(x)$

```python
x, y = sp.symbols('x y')
f = sp.Function('f')(x)
```

```python
diff_eq = sp.Eq(x * f.diff(x) - f, x**2 * sp.sin(x))
diff_eq
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
C1 = sp.Symbol('C1')
solution.subs(C1, 1)
```

```python
eq_direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
eq_direction_field
```

```python
a, b = -31, 31
x_values = np.linspace(a, b, 1000)
consts = [[(C1, i)] for i in range(-10, 11, 2)]
figure1, axes1 = plt.subplots(2, 1, layout='constrained', figsize=(8, 14))
plot_function(solution, x, diff_eq, x_values, consts, axes1[0])
c, d = a // 2, b // 2
plot_direction_field(x, y, eq_direction_field, (c, d, c, d, 1), axes1[1])
```

# $y'' - 2y' + y = 0$

```python
diff_eq2 = sp.Eq(f.diff(x, 2) - 2 * f.diff(x) + f, 0)
diff_eq2
```

```python
solution2 = sp.dsolve(diff_eq2, f)
solution2
```

```python
C2 = sp.Symbol('C2')
x_values = np.linspace(25, 30, 1000)
consts = [[(C1, np.random.randint(10)), (C2, np.random.randint(-10, 10))] for _ in range(15)]
figure2, axes2 = plt.subplots(1, 1, figsize=(8, 10))
plot_function(solution2, x, diff_eq2, x_values, consts, axes2)
```
