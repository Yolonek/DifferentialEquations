```python
from CommonFunctions import *
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
```

# $\frac{dy}{dx} = y^2$

```python
x, y = sp.symbols('x y')
f = sp.Function('f')(x)
```

```python
diff_eq = sp.Eq(f.diff(x), f**2)
diff_eq.subs(f, y)
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
C1 = sp.Symbol('C1')
a, b = -5, 5
x_values = np.linspace(a, b, 500)
consts = [[(C1, i)] for i in range(-5, 6, 2)]
figure, axes = plt.subplots(2, 1, figsize=(8, 14), layout='constrained')
plot_function(solution, x, diff_eq, x_values, consts, axes[0])
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes[1])
```

# $\frac{dy}{dx} = \sin{x} \cos{y}$

```python
diff_eq = sp.Eq(f.diff(x), sp.sin(x) * sp.cos(f))
diff_eq
```

```python
solution = sp.dsolve(diff_eq, f)
solution[0]
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
a, b = -5, 5
x_values = np.linspace(a, b, 500)
consts = [[(C1, -10)], [(C1, -5)], [(C1, -1)]]
figure3, axes3 = plt.subplots(1, 1, figsize=(8, 10), layout='constrained')
plot_function(solution[0], x, diff_eq, x_values, consts, axes3, title=False)
plot_function(solution[1], x, diff_eq, x_values, consts, axes3, title=False)
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes3, title=False)
```

# $\frac{dy}{dx} = x$

```python
diff_eq = sp.Eq(f.diff(x), x)
diff_eq
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
direction_field = diff_eq.rhs
direction_field
```

```python
consts = [[(C1, i)] for i in range(-5, 5, 1)]
figure4, axes4 = plt.subplots(1, 1, figsize=(8, 8), layout='constrained')
plot_function(solution, x, diff_eq, x_values, consts, axes4, legend=False, threshold=(a, b))
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes4, title=False)
```
