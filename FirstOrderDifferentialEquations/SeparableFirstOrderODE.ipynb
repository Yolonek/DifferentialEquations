# Separable Equations
DE is separable, when:
$$\frac{dy}{dx} = g(x)h(y)$$

```python
import sys
sys.path.append('..')
```

```python
from CommonFunctions import *
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
```

```python
x, y = sp.symbols('x y')
f = sp.Function('f')(x)
```

# $\frac{dy}{dx} = \frac{y}{1 + x}$

```python
diff_eq = sp.Eq(f.diff(x), f / (1 + x))
diff_eq.subs(f, y)
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
C1 = sp.Symbol('C1')
a, b = -5, 5
x_values = np.linspace(a, b, 500)
consts = [[(C1, i)] for i in range(-5, 6, 2)]
figure, axes = plt.subplots(1, 1, figsize=(8, 8), layout='constrained')
plot_function(solution, x, diff_eq, x_values, consts, axes, threshold=(a, b), legend=False)
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes, title=False)
axes.grid()
```

# $\frac{dy}{dx} = y^2 - 4$

```python
diff_eq = sp.Eq(f.diff(x), f**2 - 4)
diff_eq.subs(f, y)
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
consts = [[(C1, i)] for i in range(-5, 6, 2)]
figure2, axes2 = plt.subplots(1, 1, figsize=(8, 8), layout='constrained')
plot_function(solution, x, diff_eq, x_values, consts, axes2, threshold=(a, b), legend=False)
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes2, title=False)
initial_conditions = np.arange(-3, 3, 1)
plot_numeric_solution(direction_field.subs(y, f), x, f, initial_conditions, (a, b), 100, axes2, legend=False, title=True, threshold=(a, b))
```

# $(e^{2y} - y) \cos{x} \frac{dy}{dx} = e^y \sin{2x}$

```python
diff_eq = sp.Eq((sp.exp(2 * f) - f) * sp.cos(x) * f.diff(x), sp.exp(f) * sp.sin(2 * x))
diff_eq.subs(f, y)
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
figure3, axes3 = plt.subplots(1, 1, figsize=(8, 8), layout='constrained')
initial_conditions = np.arange(-4, 4, .5)
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes3, title=True)
plot_numeric_solution(direction_field.subs(y, f), x, f, initial_conditions, (a, b), 100, axes3, legend=False, title=True, threshold=(a, b))
```
