# Linear differential equations

A first order DE is linear, if:

$$ a_1(x) \frac{dy}{dx} + a_0(x)y = g(x) $$

```python
import sys
sys.path.append('..')
```

```python
from CommonFunctions import *
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x, y = sp.symbols('x y')
f = sp.Function('f')(x)
```

# $\frac{dx}{dy} - 3y = 0$

```python
diff_eq = sp.Eq(f.diff(x) - 3*f, 0)
diff_eq.subs(f, y)
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
C1 = sp.Symbol('C1')
a, b = -5, 5
x_values = np.linspace(a, b, 500)
consts = [[(C1, i)] for i in range(-5, 6, 2)]
figure, axes = plt.subplots(1, 1, figsize=(8, 8), layout='constrained')
plot_many_functions(solution, x, diff_eq, x_values, consts, axes, threshold=(a, b), legend=False)
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes, title=False)
```

# $x \frac{dy}{dx} - 4y = x^6e^x$

```python
diff_eq = sp.Eq(x * f.diff(x) - 4*f, x**6 * sp.exp(x))
diff_eq
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
a, b = -5, 5
x_values = np.linspace(a, b, 500)
consts = [[(C1, 1)], [(C1, -1)], [(C1, -2)]]
figure, axes = plt.subplots(1, 1, figsize=(8, 8), layout='constrained')
plot_many_functions(solution, x, diff_eq, x_values, consts, axes, threshold=(a, b), legend=True)
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes, title=False)
```

# $(x^2 - 9) \frac{dy}{dx} + xy = 0$

```python
diff_eq = sp.Eq((x**2 - 9) * f.diff(x) + x*f, 0)
diff_eq
```

```python
solution = sp.dsolve(diff_eq, f)
solution
```

```python
direction_field = sp.solve(diff_eq, f.diff(x))[0].subs(f, y)
direction_field
```

```python
a, b = -5, 5
x_values = np.linspace(a, b, 500)
consts = [[(C1, i)] for i in range(-5, 6, 2)]
figure, axes = plt.subplots(1, 1, figsize=(8, 8), layout='constrained')
plot_many_functions(solution, x, diff_eq, x_values, consts, axes, threshold=(a, b), legend=False)
plot_direction_field(x, y, direction_field, (a, b, a, b, .3), axes, title=False)

c, d = -2.99, 2.99
initial_x = 0
initial_conditions = np.arange(-4, 5, 1)
plot_numeric_solutions(direction_field.subs(y, f), x, f, initial_conditions, initial_x, (c, d), 100, axes, legend=False, title=False, threshold=(a, b))
```
